```python
import csv
import statistics
import pandas as pd
import random
import plotly.figure_factory as ff

df = pd.read_csv("medium_data.csv")
data = df["claps"].tolist()

def random_set_mean(counter):
  dataset = []

  for i in range(0,counter):
    cl = random.randint(0,len(data))
    value = data[cl]
    dataset.append(value)
  
  temp_mean = statistics.mean(dataset)
  return temp_mean

def show_fig(mean_list):
  df = mean_list

  fig = ff.create_distplot([df],["claps"],show_hist = False)
  fig.show()

def setup():
  mean_list = []

  for i in range(0,100):
    set_mean = random_set_mean(10)
    mean_list.append(set_mean)

  show_fig(mean_list)
  print(statistics.mean(mean_list))

setup()
```
